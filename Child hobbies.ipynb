Factors Influencing Hobbies

Olympiad Participation vs Hobbies

```python
# Count the number of students who participated in Olympiads and their predicted hobbies
olympiad_hobby = pd.crosstab(data['Olympiad_Participation'], data['Predicted Hobby'])
print(olympiad_hobby)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Count the number of students who participated in Olympiads and their predicted hobbies
olympiad_hobby = pd.crosstab(data['Olympiad_Participation'], data['Predicted Hobby'])

# Plotting the results
olympiad_hobby.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Olympiad Participation vs. Predicted Hobbies')
plt.ylabel('Number of Students')
plt.xlabel('Olympiad Participation (Yes/No)')
plt.legend(title='Predicted Hobby', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Sports Time vs Hobbies

```python
# Group the data by Predicted Hobby and calculate the average time spent on sports
sports_time_hobby = data.groupby('Predicted Hobby')['Time_sprt'].mean()
print(sports_time_hobby)
```

```python
# Group the data by Predicted Hobby and calculate the average time spent on sports
sports_time_hobby = data.groupby('Predicted Hobby')['Time_sprt'].mean()

# Plotting the average sports time for each predicted hobby
sports_time_hobby.plot(kind='bar', color='skyblue', figsize=(10, 6))
plt.title('Average Sports Time vs. Predicted Hobbies')
plt.ylabel('Average Time Spent on Sports (hours)')
plt.xlabel('Predicted Hobby')
plt.tight_layout()
plt.show()
```

Arts Time vs Hobbies

```python
# Group the data by Predicted Hobby and calculate the average time spent on arts
arts_time_hobby = data.groupby('Predicted Hobby')['Time_art'].mean()
print(arts_time_hobby)
```

```python
# Group the data by Predicted Hobby and calculate the average time spent on arts
arts_time_hobby = data.groupby('Predicted Hobby')['Time_art'].mean()

# Plotting the average arts time for each predicted hobby
arts_time_hobby.plot(kind='bar', color='lightgreen', figsize=(10, 6))
plt.title('Average Arts Time vs. Predicted Hobbies')
plt.ylabel('Average Time Spent on Arts (hours)')
plt.xlabel('Predicted Hobby')
plt.tight_layout()
plt.show()
```

Medals vs Non-Medals

```python
# Compare students who won medals and those who didn't
medal_hobby = pd.crosstab(data['Medals'], data['Predicted Hobby'])
print(medal_hobby)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cross-tabulate the number of students who won medals and their predicted hobbies
medal_hobby = pd.crosstab(data['Medals'], data['Predicted Hobby'])

# Plot Pie Charts for students who won and didn't win medals
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

# Pie chart for students who didn't win medals
medal_hobby.loc['No'].plot(kind='pie', ax=axes[0], autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'], legend=False)
axes[0].set_title('Hobbies of Students Who Didn\'t Win Medals')
axes[0].set_ylabel('')  # Hide y-label

# Pie chart for students who won medals
medal_hobby.loc['Yes'].plot(kind='pie', ax=axes[1], autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'], legend=False)
axes[1].set_title('Hobbies of Students Who Won Medals')
axes[1].set_ylabel('')  # Hide y-label

# Adjust layout
plt.tight_layout()
plt.show()
```
